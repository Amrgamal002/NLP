Bag Of Words


```python
import pandas as pd

df=pd.read_csv(r'C:\Users\Amr\Downloads\spam.csv')
df.head()
```

```python
df.Category.value_counts()
```

```python
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam'else 0)
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.Message,df.spam,test_size=.2,random_state=22)
type(x_train.values)
```

```python
x_train.shape
```

Create bag of words representation using CountVectorizer


```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer()
x_train_cv=cv.fit_transform(x_train)
x_test_cv=cv.transform(x_test)
x_train_nlp=x_train_cv.toarray()
x_train_nlp.shape

```

```python
print(cv.vocabulary_)
```

```python
import numpy as np
np.where(x_train_nlp[0]!=0)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train_nlp, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test_cv)

print(classification_report(y_test, y_pred))
```

Train the model using sklearn pipeline and reduce number of lines of code


```python
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)

print(classification_report(y_test, y_pred))
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# مجموعة النصوص التدريبية كمثال
corpus = [
    "I'm from Egypt.",
    "I work as a ai engineer.",
    "The field of artificial intelligence is developing in Egypt."
]

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

# يمكنك الحصول على متجه TF-IDF لكل وثيقة في المصفوفة
# tfidf_matrix هي مصفوفة sparse (تحتوي على الكثير من الأصفار)
print("TF-IDF:shape", tfidf_matrix.shape)
print( tfidf_vectorizer.get_feature_names_out())
print( tfidf_matrix[1])


tfidf_vectors = tfidf_matrix.toarray()
print( tfidf_vectors[0])


```
